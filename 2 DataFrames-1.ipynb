```python
#### Broadcasting Operations on DataFrames
```

```python
# It will add 5 to each item
nba['Age'].add(5)
nba['Age'] + 5 #output same as above
```

```python
#sub
nba['Salary'].sub(5000)
nba['Salary'] - 5000 #output same as above
```

```python
#multipy

#.453592 convert weight to pounds to kg

nba['Weight'].mul(.453592)

#add new column with updated series
nba['Weight_In_KG'] = nba['Weight'].mul(.453592)
```

```python
nba.head()
```

#### Drop DataFrame Rows with Null Values

```python
# it will remove all rows, where any of column have NaN value
nba.dropna()
```

```python
# Remove only where all columns have NaN value (ROW Wise)
nba.dropna(how="all")

# nba.dropna(how="all", inplace=True)
```

```python
nba.tail(3)
```

```python
# Drop Column Wise

nba.dropna(axis=1)
nba.dropna(axis="columns")
nba.tail()
```

```python
# Remove for specific columns

nba.dropna(subset=['Salary'])
nba.dropna(subset=['Salary','College'])
```

```python
#### Change Data Type
# Chnage a datatype also effect memory usage
# like int uses less memory according to float
```

```python
# fill NaN as 0 in salary to avoid errors
nba['Salary'].fillna(0, inplace=True)
nba['Salary']
```

```python
# Change datatype
nba['Salary'] = nba['Salary'].astype('int')


nba['Salary'].head()
```

```python
# Category datatype
# Its like a enum data type, it uses to store uniqe or similar values
# Here we have Position column that can be a "category" datatype

# to check uniqueness
nba['Position'].nunique()
```

```python
nba['Position'] = nba['Position'].astype('category')

# It will show same output,
```

```python
#### Sort a DataFrame with the sort_values
```

```python
nba.sort_values("Salary")

nba.sort_values("Salary", ascending=False)
# nba.sort_values("Salary", ascending=False, inplace=True)

nba.sort_values("Salary", ascending=False, na_position="first")
# nba.sort_values("Salary", ascending=False, na_position="last")
```

```python
## Sort by muti columns
```

```python

nba.sort_values(["Team", "Name"])

# order
nba.sort_values(["Team", "Name"], ascending=False)
nba.sort_values(["Team", "Name"], ascending=[True, False])
# nba.sort_values(["Team", "Name"], ascending=[True, False], inplace=True)
```

#### rank method

```python
nba['Salary'] = nba['Salary'].fillna(0).astype('int')

nba['Salary_Rank'] = nba['Salary'].rank(ascending = False).astype('int')
```
