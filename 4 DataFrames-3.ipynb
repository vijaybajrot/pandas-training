#### Rename Index Labels or Columns in a DataFrame

```python
# changing by default row wise
bond.rename(mapper={'Thunderball': 'Thunder ball', 'Diamonds Are Forever': 'Diamonds Are Forever Again'})
```

```python
bond.rename(mapper={'Thunderball': 'Thunder ball', 'Diamonds Are Forever': 'Diamonds Are Forever Again'}, axis=0)
```

```python
#use inplace to update
bond.rename(mapper={'Thunderball': 'Thunder ball', 'Diamonds Are Forever': 'Diamonds Are Forever Again'}, axis='rows')
#same as above
bond.rename(mapper={'Thunderball': 'Thunder ball', 'Diamonds Are Forever': 'Diamonds Are Forever Again'}, axis='index')
```

#### Change columns

```python
# 

bond.rename(mapper={'Year': 'Released Date', 'Box Office': 'Revenue'}, axis='columns')
```

```python
bond.rename(mapper={'Year': 'Released Date', 'Box Office': 'Revenue'}, axis=1)
```

```python
# using columns

bond.rename(columns={'Year': 'Released Date', 'Box Office': 'Revenue'})
```
