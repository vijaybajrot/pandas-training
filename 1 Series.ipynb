#### Reindex

```python
# It will throw error, key not exists
pokemonSeries['Digimon']
```

```python
# Bulbasaur exists so it will remain same
# It will create Digimon key with empty value

pokemonSeries.reindex(['Bulbasaur', 'Digimon'])
```

####  get Method to Retrieve a Value

```python
pokemonSeries.get(0)
pokemonSeries.get([0,5])
```

```python
pokemonSeries.get("Bulbasaur")
pokemonSeries.get('Bulbasaur', 'Ivysaur')
```

```python
# It will return None, key not exists
pokemonSeries.get('Digimon')
pokemonSeries.get(['Bulbasaur', 'Digimon'])
```

```python
# get method by default value
pokemonSeries.get('Digimon', default="NotFound")
```

```python
pokemonSeries.get(['Bulbasaur', 'Digimon'], default="Any of one not found")
```

#### The map function

```python
pokemonNames = pd.read_csv("files/pokemon.csv", usecols = ['Pokemon'], squeeze=True)
pokemonTypes = pd.read_csv("files/pokemon.csv", index_col = 'Pokemon', squeeze=True)
```

```python
pokemonNames.map(pokemonTypes)
```

```python
### using Dict
```

```python
pokemonTypesDict = pokemonTypes.to_dict()
```

```python
pokemonNames.map(pokemonTypesDict)
```
