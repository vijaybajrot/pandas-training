#### The transpose Method on a MultiIndex DataFrame

```python
bigmac2 = bigmac.copy()

bigmac2 = bigmac2.transpose()
bigmac2
```

```python
bigmac2.loc[('Price in US Dollars',), ("2010-01-01",)]
bigmac2.loc[('Price in US Dollars',), ("2010-01-01","Argentina")]


# type(bigmac2.loc[('Price in US Dollars',), ("2010-01-01",)])
```

```python
bigmac2.loc[('Price in US Dollars',), ("2010-01-01","Sri Lanka"):("2010-01-01","Ukraine")]
```

#### The .swaplevel() Method

#### The .stack() Method

```python
world = pd.read_csv("files/worldstats.csv", index_col=['country', 'year'])
```

```python
world.head()
```

```python
# shift columns to row index

world.stack()

type(world.stack())
```

```python
world2 = world.stack().to_frame()

world2
```

```python


world2.loc[('Arab World', 2015, 'Population')]
```

```python
# shift row index to column index

world2 = world2.unstack()
```

```python
# world2.index

# world2.loc[('Arab World', 2015, 'Population')]

world2
```

```python
# world2.loc[('Afghanistan',), ("Population", 1960)]


world2.index

world2.loc[('Afghanistan',), (0, "Population", 1960)]

# world2.columns
```
