```python
import pandas as pd
```

```python
week1 = pd.read_csv("files/Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv("files/Restaurant - Week 2 Sales.csv")
customers = pd.read_csv("files/Restaurant - Customers.csv")
foods = pd.read_csv("files/Restaurant - Foods.csv")
```

```python
foods
```

```python
# pd.concat Method

# concat 2 dataframes

pd.concat(objs=[week1, week2])

# rows should be 250*2 = 500
```

```python
pd.concat(objs=[week1, week2], ignore_index=True)
```

```python
# another example
# it will create multi index
sales = pd.concat(objs=[week1, week2], keys=['Week1', 'Week2'])
```

```python
#### Inner Joins
```

```python
week1.merge(week2, how='inner', on='Customer ID')
```

```python
#notice above, 155 repeats

week1[week1['Customer ID'] == 155]
```

```python
week2[week2['Customer ID'] == 155]
```

```python
week1.merge(week2, how='inner', on='Customer ID', suffixes=['_week1', '_week2'])
```

```python
week1.merge(week2, how='inner', on=['Customer ID', 'Food ID'])
```

```python
# outer join

weeksdata = week1.merge(week2, how='outer', on='Customer ID', indicator=True)
weeksdata
# len(weeksdata)
```

```python
mask = weeksdata['_merge'] != 'both'

weeksdata[mask]
```

```python
foods.head()

# week1.head()
```

```python
# left join

week1.merge(foods, how='left', on='Food ID').head()
```

```python
# Sort by Food ID
week1.merge(foods, how='left', on='Food ID', sort=True).head()
```

```python
week2.head()

customers.head()
```

```python
week2.merge(customers, how='left', left_on='Customer ID', right_on='ID').drop('ID', axis=1).head()
```

```python
#Merging by Indexes with the left_index and right_index Parameters

foods.set_index('Food ID', inplace=True)
```

```python
customers.set_index('ID', inplace=True)
```

```python
week1.merge(customers, how='left', left_on='Customer ID', right_index=True)
```

```python
week1.merge(week2, how='left', left_index=True, right_index=True).head()
```
