```python
#export to excel

excel_writer = pd.ExcelWriter("files/Baby_Names.xlsx")
```

```python
girls = babynames[babynames['Gender'] == 'FEMALE']
boys = babynames[babynames['Gender'] == 'MALE']


girls
```

```python
girls.to_excel(excel_writer, sheet_name='Girls', index=False)
```

```python
boys.to_excel(excel_writer, sheet_name='Boys', index=False, columns=['Year of Birth', 'Gender', 'Ethnicity'])
```

```python
excel_writer.save()
```
